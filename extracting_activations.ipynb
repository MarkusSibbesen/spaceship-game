```python
from hook_manager import HookManager
from data_handling import load_sentiment_data
from transformers import AutoModelForCausalLM, AutoTokenizer

import torch
import numpy as np
from collections import defaultdict
```

```python
%load_ext autoreload
%autoreload 2
```

```python
data = load_sentiment_data('data/train.tsv')
```

```python
model_name = "roneneldan/TinyStories-33M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
activations = dict()
labels = []

with HookManager(model) as hook_manager:
    for layer in range(model.config.num_layers):
        activations[layer] = hook_manager.attach_residstream_hook(layer)

    for idx, datapoint in enumerate(data): 
        text = datapoint['text']
        label = datapoint['label']

        tokenized = tokenizer(text, return_tensors='pt')
        model(**tokenized)

        for _ in tokenized.input_ids[0]:
            labels.append(int(label))

        if idx > 1000:
            break
        

labels = np.array(labels)

for layer in activations:
    activations[layer] = torch.concat(activations[layer])
```

```python
for layer in activations:
    for label in np.unique(labels):
        torch.save(
            activations[layer][labels == label].mean(dim=0), 
            f'steering_vectors/raw_means/layer{layer}_label{label}.pt'
        )
```
